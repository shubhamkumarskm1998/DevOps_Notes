# 🧪 What is Flask?

**Flask** is a lightweight Python web framework used to build web applications and APIs.

- Minimal and simple (micro-framework)
- Does not enforce strict structure
- Easy to learn and flexible
- Commonly used for APIs, small apps, and microservices

---

# 🏗️ Background of Flask

- Created by **Armin Ronacher**
- First released in **2010**
- Inspired by the **Werkzeug** (WSGI toolkit) and **Jinja2** (template engine)
- Developed as part of the **Pocoo project**

---

# ⚙️ Why Flask is called a Micro Framework?

Flask is called a micro-framework because:

- It does **not include**:
  - Database layer (ORM)
  - Form validation
  - Authentication system

- You can **add these features manually** using extensions

👉 Example:
- Database → SQLAlchemy
- Authentication → Flask-Login

---

# 🚀 Features of Flask

- Built-in development server
- URL routing
- RESTful request handling
- Jinja2 templating engine
- Lightweight and extensible
- Easy integration with databases and tools

---

# 🔄 Flask Architecture

- Based on **WSGI (Web Server Gateway Interface)**
- Uses:
  - **Werkzeug** → handles HTTP requests/responses
  - **Jinja2** → renders HTML templates

---

# 💡 Why Flask is Popular?

- Beginner-friendly
- Highly flexible
- Great for APIs and microservices
- Widely used in startups and DevOps workflows

---

# 📦 Where Flask is Used?

- REST APIs
- Machine Learning model deployment
- Microservices
- Backend for web apps

# 📊 Types of Databases

## 4 Main Types of Databases

### 1. Relational / SQL Databases
- Store data in tables (rows & columns)
- Use SQL (Structured Query Language)

**Examples:**
- MySQL
- PostgreSQL
- Oracle
- SQL Server

---

### 2. NoSQL / Document Databases
- Store data in JSON-like documents
- Flexible schema

**Examples:**
- MongoDB
- CouchDB
- Firebase Firestore

---

### 3. Graph Databases
- Store data as nodes and relationships
- Best for connected data

**Examples:**
- Neo4j
- Amazon Neptune

---

### 4. Key-Value Databases
- Store data as key-value pairs
- Very fast lookups

**Examples:**
- Redis
- DynamoDB
- Riak

---

## 5. Blob Stores

- Store large unstructured data (images, videos, files)

**Examples:**
- Amazon S3
- Azure Blob Storage
- Google Cloud Storage

# 🍃 Introduction to MongoDB

## What is MongoDB?

MongoDB is a **NoSQL (non-relational) database** that stores data in **JSON-like documents (BSON)** instead of tables.

- Data is stored as **documents**
- Documents are grouped into **collections**
- No fixed schema (flexible structure)

---

## ❓ Why MongoDB?

Traditional databases (SQL) have limitations:

- Fixed schema (hard to change)
- Complex joins
- Not ideal for large-scale, fast-changing data

👉 MongoDB solves this by being:
- Flexible
- Scalable
- Developer-friendly

---

## 🚀 Why do we need MongoDB?

MongoDB is useful when:

- Data structure **keeps changing**
- You need **high performance**
- You handle **large-scale data (Big Data)**
- You build **modern apps (APIs, microservices)**

---

## ⭐ What makes MongoDB special?

### 1. 📄 Document-Based Storage
```json
{
  "name": "Shubham",
  "age": 25,
  "skills": ["Python", "DevOps"]
}
```
👉 No tables, no rows — just flexible documents

---

### 2. 🧩 Schema-less (Flexible)
- You don’t need to define structure beforehand
- Each document can be different

---

### 3. ⚡ High Performance
- Faster reads/writes compared to many SQL systems
- No complex joins

---

### 4. 📈 Horizontal Scaling
- Can scale across multiple servers
- Uses **sharding**

---

### 5. 🔄 Built for Modern Apps
- Works great with:
  - REST APIs
  - Microservices
  - Real-time apps

---

## 🆚 MongoDB vs SQL (Quick View)

| Feature        | MongoDB        | SQL Database     |
|----------------|----------------|------------------|
| Structure      | Document       | Table            |
| Schema         | Flexible       | Fixed            |
| Scaling        | Horizontal     | Vertical         |
| Joins          | Limited        | Strong           |
| Use Case       | Modern apps    | Structured data  |

---

## 🎯 When to Use MongoDB?

- Dynamic or evolving data
- Real-time applications
- Large-scale distributed systems
- JSON-heavy APIs

# 🧩 Microservices - Concept

## What are Microservices?

Microservices is an **architecture style** where an application is built as a collection of **small, independent services**.

👉 Each service:
- Does **one specific task**
- Runs independently
- Communicates via APIs (HTTP/REST)

---

## 🧠 Simple Idea

Instead of one big app:

```
Monolith App = Everything in ONE codebase
```

We split it into:

```
Microservices = Small independent services
```

---

## 🏗️ Example (E-commerce App)

Instead of one app:

- User Service → login/signup
- Product Service → product data
- Order Service → orders
- Payment Service → payments

👉 Each runs separately

---

## 🔄 How Microservices Work

```
Client → API Gateway → Microservices → Database
```

- Client sends request
- API Gateway routes it
- Services process independently

---

## 🔑 Key Characteristics

- Independent deployment
- Loosely coupled
- Each service has its own database
- Communicate via APIs
- Can use different technologies

---

## ⚖️ Microservices vs Monolith

| Feature        | Monolith            | Microservices        |
|----------------|--------------------|---------------------|
| Structure      | Single app         | Multiple services   |
| Deployment     | One unit           | Independent         |
| Scaling        | Whole app          | Per service         |
| Complexity     | Simple initially   | Complex             |
| Flexibility    | Low                | High                |

---

## 🚀 Advantages

- Easy to scale specific services
- Faster development
- Fault isolation (one service fails, others run)
- Tech flexibility

---

## ⚠️ Disadvantages

- More complex system
- Requires DevOps setup
- Network communication overhead
- Harder debugging

---

## 🛠️ Technologies Used

- Backend → Flask, Node.js, Spring Boot
- Communication → REST APIs, gRPC
- Containers → Docker
- Orchestration → Kubernetes
- Messaging → Kafka, RabbitMQ

---

## 🎯 When to Use Microservices?

- Large applications
- Multiple teams working
- Need for scalability
- Frequent deployments

---
