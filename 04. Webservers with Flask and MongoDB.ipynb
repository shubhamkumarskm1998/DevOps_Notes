# 🌐 What is a Web Framework?

A **Web Framework** is a tool or library that helps developers build web applications easily by providing:
- Pre-written code
- Structure for development
- Tools for handling requests, routing, and responses

---

# 🧪 What is Flask?

**Flask** is a lightweight Python web framework used to build web applications and APIs.

- Minimal and simple (micro-framework)
- Does not enforce strict structure
- Easy to learn and flexible
- Commonly used for APIs, small apps, and microservices

---

# 🏗️ Background of Flask

- Created by **Armin Ronacher**
- First released in **2010**
- Inspired by the **Werkzeug** (WSGI toolkit) and **Jinja2** (template engine)
- Developed as part of the **Pocoo project**

---

# ⚙️ Why Flask is called a Micro Framework?

Flask is called a micro-framework because:

- It does **not include**:
  - Database layer (ORM)
  - Form validation
  - Authentication system

- You can **add these features manually** using extensions

👉 Example:
- Database → SQLAlchemy
- Authentication → Flask-Login

---

# 🚀 Features of Flask

- Built-in development server
- URL routing
- RESTful request handling
- Jinja2 templating engine
- Lightweight and extensible
- Easy integration with databases and tools

---

# 🔄 Flask Architecture

- Based on **WSGI (Web Server Gateway Interface)**
- Uses:
  - **Werkzeug** → handles HTTP requests/responses
  - **Jinja2** → renders HTML templates

---

# 💡 Why Flask is Popular?

- Beginner-friendly
- Highly flexible
- Great for APIs and microservices
- Widely used in startups and DevOps workflows

---

# 📦 Where Flask is Used?

- REST APIs
- Machine Learning model deployment
- Microservices
- Backend for web apps

````python
from flask import Flask

app = Flask(__name__)

@app.route("/home/")
def home():
    return "Hello, Welcome to Home page"

if __name__ == "__main__":
    app.run(debug=True,use_reloader=False)
    ##use_reloader is necessary if we want to run this code in Notebook
````

## 🔑 Important Terms

### 1. Client (Browser)
- Chrome, Edge, Postman
- Sends request to server

Example:
```
http://localhost:5000/
```

---

### 2. HTTP Request

Types:
- GET → get data
- POST → send data
- PUT → update data
- DELETE → delete data

---

### 3. Flask App

```python
from flask import Flask
app = Flask(__name__)
```

- Creates your application/server

---

### 4. Routing

```python
@app.route("/")
```

- Maps URL → function

```python
@app.route("/")
def home():
    return "Hello"
```

---

### 5. View Function

```python
def home():
    return "Hello"
```

- Runs logic
- Returns response

---

### 6. Request → Response

```python
return "Hello, Flask!"
```

Response can be:
- Text
- HTML
- JSON

---

### 7. JSON (Important)

```json
{
  "name": "Shubham",
  "role": "DevOps"
}
```

- Standard format for APIs

---

### 8. Templates (Jinja2)

```python
return render_template("index.html")
```

- Used to render HTML
- Allows dynamic content

---

### 9. Werkzeug

- Internal tool used by Flask
- Handles requests and responses

---

### 10. WSGI

- Web Server Gateway Interface

```
Web Server ↔ Flask App
```

---

### 11. Development Server

```python
app.run()
```

- Runs app locally

URL:
```
http://127.0.0.1:5000
```

---

### 12. Debug Mode

```python
app.run(debug=True)
```

- Auto reload
- Shows detailed errors

---

### 13. Extensions

Flask is minimal → add features:

- Database → SQLAlchemy
- Login → Flask-Login
- APIs → Flask-RESTful

---

## 🔄 Full Flow

1. User opens browser
2. Sends request
3. Flask receives request
4. Route matches function
5. Function runs
6. Response returned
7. Browser displays result
