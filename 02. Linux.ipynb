# 🐧 Introduction to Linux

## 1. What is Linux, and Who Created it?
Linux is an **open-source operating system** used to manage hardware and software.

- Created by **Linus Torvalds** in 1991
- Based on Unix-like design
- Used in servers, cloud, and development environments

---

## 2. GNU and the GPL

- **GNU Project**: Started to create free and open-source software
- **GPL (General Public License)**:
  - Allows users to use, modify, and share software freely
  - Ensures software remains open-source

👉 Linux + GNU tools = complete OS

---

## 3. Uses of Linux

- Web servers 🌐
- Cloud computing ☁️
- DevOps & automation ⚙️
- Android OS 📱
- Supercomputers 🖥️

---

## 4. Advantages and Disadvantages

### ✅ Advantages
- Free and open-source
- Secure and stable
- Highly customizable
- Lightweight

### ❌ Disadvantages
- Steeper learning curve
- Less support for some software (e.g., certain games/apps)
- Hardware compatibility issues (rare now)

```bash
ls        # list files
cd        # change directory
pwd       # present working directory
mkdir     # create folder
rm        # delete file/folder
cp        # copy files
mv        # move/rename files

# 🐧 Linux Notes

---

# 📜 History

- The history of Linux began with **Unix in 1969**
- Unix was created at **Bell Labs** with goals:
  - Simplicity
  - Reusable code
  - Written in C (instead of assembly)

- Linux development started in **1991**
- **Linus Torvalds** wanted to create a free version of Unix
- By **1993**, there were ~12,000 Linux users

---

# ⚙️ The Kernel

- Linux is **not a full Operating System**
- Linux is a **kernel**

## What is a Kernel?
A kernel is a program that:
- Manages hardware
- Allocates system resources (CPU, memory, devices)

👉 Linux distributions = **Linux kernel + GNU tools**

---

# 📜 What is GNU GPL?

**GNU GPL (General Public License)**

- Ensures software remains **free and open-source**
- Based on **Copyleft** (freedom to use, modify, share)
- Modified versions must also be open-source
- Linux is released under GPL

---

# 🚀 Why Use Linux?

## As a Server Platform
- High performance
- Low cost (free)
- Very stable

## For Developers
- Huge number of free tools
- Great for programming and automation

## For Desktop
- Flexible and customizable
- Fun to use

---

# ⚠️ Challenges of Linux

- Steeper learning curve than Windows
- Less user-friendly for beginners
- Some software compatibility issues
- Manual configuration sometimes required

---

# 🧩 Popular Linux Distributions

- Ubuntu
- Kali Linux
- Debian
- CentOS
- Alpine
- Arch Linux
- Fedora

---

# 📁 Linux File System Basics

Linux follows a **hierarchical structure**:

👉 Everything starts from root:


## Important Directories

- `/boot` → Kernel and boot files
- `/bin` → Basic system commands
- `/dev` → Device files
- `/etc` → Configuration files
- `/home` → User directories
- `/lib` → System libraries
- `/mnt` → Mount points
- `/proc` → System/process info
- `/root` → Root user home
- `/sbin` → System binaries
- `/usr` → User programs
- `/var` → Logs and variable data

---

# 🎯 Summary

- Linux is a **kernel**, not a full OS
- Built using **GNU + Linux kernel**
- Open-source under **GPL license**
- Widely used in servers, cloud, and DevOps

# 🐧 Linux Commands Cheat Sheet

## 📁 File & Directory Commands

- `ls` → List directories and files
- `pwd` → Show current directory
- `cd` → Navigate between directories
- `mkdir` → Create a new directory
- `mv` → Move or rename files
- `cp` → Copy files
- `rm` → Delete files or directories
- `touch` → Create empty file

---

## 🔗 Links & Display

- `ln` → Create symbolic links
- `cat` → Display file content
- `clear` → Clear terminal screen

---

## 📖 Help & Info

- `man` → Show manual pages
- `uname` → System information
- `whoami` → Current user

---

## 📦 Compression & Search

- `tar` → Compress/extract files
- `zip` → Zip files
- `unzip` → Unzip files
- `grep` → Search text in files

---

## 📊 File Viewing

- `head` → Show first lines of file
- `tail` → Show last lines of file

---

## 🌍 Environment & Variables

- `export` → Set environment variables

---

## 🔐 Remote & Services

- `ssh` → Secure remote login
- `service` → Start/stop services

---

## ⚙️ Process Management

- `ps` → Show running processes
- `kill` → Kill process by ID
- `top` → Live process monitoring

---

## 💾 Disk & Permissions

- `df` → Disk usage
- `chmod` → Change file permissions

---

## 🌐 Networking

- `ifconfig` → Network info (IP address)
- `wget` → Download files from internet

---

## 🔥 Firewall

- `ufw` → Firewall management

---

## 📦 Package Management

- `apt`, `yum`, `pacman` → Install/manage packages

---

## 🔑 Permissions

- `sudo` → Run command as administrator

---

## ⚡ Shortcuts

- `alias` → Create shortcuts for commands

---

## 🔍 Command Info

- `whereis` → Locate command binaries
- `whatis` → Short description of a command

---

## 🎯 Summary

These commands are essential for:
- File management
- Process handling
- Networking
- System administration
- DevOps workflows
