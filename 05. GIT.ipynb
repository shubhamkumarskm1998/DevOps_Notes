# 📦 Git & GitHub Basics

## ❓ What is Git?

- Git is a **version control system**
- Helps track changes in code
- Allows multiple developers to work together

---

## 🌐 What is GitHub?

- GitHub is a **cloud platform** to host Git repositories
- Used for:
  - Storing code
  - Collaboration
  - Code sharing

---

## 🤔 Why use Git & GitHub?

- Track code changes
- Collaborate with teams
- Maintain code history
- Backup code online
- Work on multiple features using branches

```bash

- Sets remote `origin`
- Creates local copy

---

```bash

```

- Downloads updates
- Does NOT modify working code

---

## Pull Changes

```bash
git pull
```

- Fetch + merge
- Updates local branch

# 🕒 Git History & Undo Operations

---

## 🔙 Removing Changes from Staging

### Unstage a file

```bash
git restore --staged file.txt
```

OR (older command)

```bash
git reset file.txt
```

---

## 📜 Viewing Project History

```bash
git log
```

👉 Shows full commit history

---

### Short view

```bash
git log --oneline
```

---

### Graph view (very useful)

```bash
git log --oneline --graph --all
```

---

## ✏️ Making More Commits

```bash
git add .
git commit -m "new changes"
```

👉 You can commit as many times as needed

---

## ❌ Removing a Commit

### 1. Remove last commit (keep changes)

```bash
git reset --soft HEAD~1
```

---

### 2. Remove last commit (delete changes)

```bash
git reset --hard HEAD~1
```

---

### 3. Remove specific commit (safe way)

```bash
git revert <commit_id>
```

👉 Creates a new commit that undoes changes

---

## ⚠️ Important

- `reset --hard` → deletes changes permanently
- `revert` → safer for shared code

---

## 🚀 Example Workflow

```bash
git log --oneline
git reset --soft HEAD~1
git add .
git commit -m "fixed commit"
```

---

## 🎯 One-Line Summary

Unstage → View History → Commit → Reset/Revert

# 🌿 Git Branches

---

## ❓ What are Branches?

- A **branch** is a separate line of development
- Allows you to work on features **without affecting main code**

👉 Default branch = `main` (or `master`)

---

## 🎯 Use of Branches

- Develop new features
- Fix bugs
- Experiment safely
- Work in teams without conflicts

---

## 🌱 Create and Switch Branch

### Create branch

```bash
git branch feature
```

---

### Switch branch

```bash
git checkout feature
```

OR (modern)

```bash
git switch feature
```

---

### Create + Switch

```bash
git checkout -b feature
```

OR

```bash
git switch -c feature
```

---

## 🔀 Merge Branch to Main

```bash
git checkout main
git merge feature
```

👉 Combines feature branch into main

---

## 📤 Push Changes to Remote (Master/Main)

```bash
git push origin main
```

👉 Push local main branch to remote

---

## 📤 Push New Branch

```bash
git push origin feature
```

---

## ⚠️ Best Practice

- Always create branch for new work
- Don’t work directly on `main`
- Merge after testing

---

## 🚀 Example Workflow

```bash
git checkout -b feature
# make changes
git add .
git commit -m "new feature"
git checkout main
git merge feature
git push origin main
```

---

## 🎯 One-Line Summary

Branch → Work → Merge → Push

- Create branch
- Push branch

- Review
- Merge

# 🔁 Pull Request (PR)

---

## ❓ What is a Pull Request?

- A **Pull Request (PR)** is a request to merge changes from one branch into another
- Commonly used in GitHub

👉 Example:
```
feature → main
```

---

## 🎯 Why use Pull Requests?

- Code review
- Team collaboration
- Maintain code quality
- Track changes before merging

---

## 🚀 Creating Your First Pull Request

### Step 1: Create a branch

```bash
git checkout -b feature
```

---

### Step 2: Make changes

```bash
git add .
git commit -m "new feature"
```

---

### Step 3: Push branch

```bash
git push origin feature
```

- Click **Compare & Pull Request**
- Add title & description
- Click **Create Pull Request**

## ❌ Removing a Commit from PR (Force Push)

### Step 1: Reset commit

```bash
git reset --soft HEAD~1
```

OR (delete completely)

```bash
git reset --hard HEAD~1
```

### Step 2: Force push

```bash

### ⚔️ What is a Git Conflict?

👉 Happens when:

Same file
Same lines
Changed in two branches

Git doesn’t know which version to keep ❌

````python
"""<<<<<<< HEAD
print("Hello from main")
=======
print("Hello from feature")
>>>>>>> feature"""
````

### Two ways to fix this
    - 1. Remove the code manually
    - 2. rebase -> git rebase main
                - git add file.py
                - git rebase --continue


# 🍴 Fork, Clone & Upstream (GitHub Workflow)

---

## ❓ Why Fork?

- To **copy someone else's repository** into your GitHub account
- Used for:
  - Contributing to open-source
  - Making changes without affecting original repo

---

## 🔧 How to Fork?

1. Go to GitHub repository
2. Click **Fork** button (top right)
3. Repo will be copied to your account

---

## 📥 Cloning the Forked Project

```bash
git clone https://github.com/your-username/repo.git
```

👉 This downloads your fork to local system

---

## 🔗 What is Upstream?

- **Upstream** = original repository (not your fork)
- Used to get latest updates from original project

---

## ➕ Adding Upstream to Local Repo

```bash
git remote add upstream https://github.com/original-owner/repo.git
```

---

## 🔍 Verify remotes

```bash
git remote -v
```

Example:
```text
origin    https://github.com/your-username/repo.git
upstream  https://github.com/original-owner/repo.git
```

---

## 🔄 Sync with Upstream

```bash
git fetch upstream
git merge upstream/main
```

OR

```bash
git pull upstream main
```

---

## 🚀 Complete Fork Workflow

```bash
# Fork repo (GitHub UI)

git clone your-fork-url
cd repo

git remote add upstream original-repo-url

git fetch upstream
git merge upstream/main

# create branch
git checkout -b feature

# make changes
git add .
git commit -m "update"

git push origin feature

# create Pull Request
```

---

## 🎯 One-Line Summary

Fork → Clone → Add Upstream → Sync → Create PR

# 🧹 Squashing & Stashing in Git

---

## 🔗 Squashing Commits

### What is Squash?

- Combine multiple commits into **one single commit**
- Used to keep **clean history**

---

### Example

```bash
git rebase -i HEAD~3
```

👉 Opens editor:

```
pick commit1
pick commit2
pick commit3
```

👉 Change to:

```
pick commit1
squash commit2
squash commit3
```

---

## 📦 Stashing Changes

### What is Stash?

- Temporarily save changes **without committing**
- Useful when switching branches

---

### Save changes

```bash
git stash
```

---

### Stash with message

```bash
git stash save "work in progress"
```

---

## 📤 Popping Stash

### Apply + remove stash

```bash
git stash pop
```

---

## 📥 Apply without removing

```bash
git stash apply
```

---

## 🧹 Clearing Stash

### Remove all stashes

```bash
git stash clear
```

---

### Remove specific stash

```bash
git stash drop stash@{0}
```

---

## 🔍 View stash list

```bash
git stash list
```

---

## 🚀 Example Workflow

```bash
git stash
git checkout main
git pull
git checkout feature
git stash pop
```

---

## 🎯 One-Line Summary

Squash = combine commits
Stash = save work temporarily

# 🔀 Rebase, Reset (Hard) & Merge Conflicts

---

## 🔄 Using the Rebase Command

### What is Rebase?

- Rebase = move your commits **on top of another branch**
- Keeps **clean, linear history**

---

### Example

```bash
git checkout feature
git rebase main
```

👉 Your feature commits will be applied after latest main

---

### Continue after conflict

```bash
git add file.txt
git rebase --continue
```

---

### Abort rebase

```bash
git rebase --abort
```

---

## 💣 Using Hard Reset

### What is `--hard`?

- Completely **removes commits and changes**
- Dangerous (data loss)

---

### Example

```bash
git reset --hard HEAD~1
```

👉 Deletes last commit permanently

---

### Reset to specific commit

```bash
git reset --hard <commit_id>
```

---

## ⚔️ Merge Conflicts

### When do conflicts happen?

- Same file edited in different branches
- Git cannot auto-merge

---

### Example Conflict

```text
<<<<<<< HEAD
your code
=======
other code
>>>>>>> feature
```

---

## 🛠️ How to Resolve Conflicts

### Step 1: Open file & fix manually

- Choose correct code
- Remove conflict markers

---

### Step 2: Stage changes

```bash
git add file.txt
```

---

### Step 3: Complete process

#### For merge:

```bash
git commit
```

#### For rebase:

```bash
git rebase --continue
```

---

## ❌ Abort if needed

```bash
git merge --abort
```

OR

```bash
git rebase --abort
```

---

## ⚠️ Best Practices

- Avoid `reset --hard` on shared branches
- Pull latest changes before working
- Resolve conflicts carefully

---

## 🎯 One-Line Summary

Rebase = clean history
Reset --hard = delete changes
Conflict = manual fix required

- A file used to **ignore files/folders from Git tracking**

👉 Place `.gitignore` in root folder
