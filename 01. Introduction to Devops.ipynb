# Introduction to DevOps

## 1. What is DevOps
DevOps is a practice that combines **Development (Dev)** and **Operations (Ops)** to improve collaboration and automate software delivery.

---

## 2. DevOps Main Objectives
- Faster delivery of applications 🚀
- Better collaboration between teams
- Automation of processes
- High reliability and scalability

---

## 3. DevOps and Software Development Life Cycle (SDLC)
DevOps integrates with SDLC stages:
- Planning → Development → Testing → Deployment → Monitoring
👉 Makes the process continuous and automated

---

## 4. Continuous Integration & Deployment (CI/CD)
- **CI (Continuous Integration):** Automatically build and test code when changes are made -> Jenkins, GITHUB ACTIONS
- **CD (Continuous Deployment/Delivery):** Automatically deploy code to production -> Jenkins, GITHUB ACTIONS

👉 Reduces manual work and errors

---

## 5. Containers and Virtual Development
- **Containers (e.g., Docker):** Package app + dependencies together
- **Virtual Machines:** Full OS-based environments

👉 Ensures consistency across environments

---

## 6. Cloud Providers
Platforms that provide infrastructure and services:
- AWS
- Azure
- Google Cloud

👉 Used for hosting, storage, and scaling applications

---

## 7. Scripting
Writing scripts (Bash, Python, etc.) to:
- Automate tasks
- Deploy applications
- Manage systems

👉 Reduces manual effort

---

## 8. Configuration Management Tools
Tools used to manage system configurations:
- Ansible
- Puppet
- Chef

👉 Ensures systems are consistent and reproducible

---

## 🎯 Summary
DevOps focuses on **automation, collaboration, and continuous delivery** to build and release software faster and more reliably.

# ☁️ What is Cloud?

Cloud computing means using **servers, storage, and services over the internet** instead of your local computer.

👉 You don’t need to buy hardware
👉 You rent resources when needed

---

## 🎯 Simple Definition
Cloud is the delivery of computing services (servers, storage, databases, networking, software) over the internet.

---

## 💡 Example
Instead of storing files on your laptop:
- You use Google Drive or AWS S3
👉 That is cloud storage

---

# 🔹 Types of Cloud Services

## 1. IaaS (Infrastructure as a Service)

👉 Provides basic infrastructure:
- Virtual machines
- Storage
- Networking

You manage:
- OS
- Applications

### Examples:
- Amazon EC2 (AWS)
- Azure Virtual Machines
- Google Compute Engine

---

## 2. PaaS (Platform as a Service)

👉 Provides platform to build applications

You don’t manage:
- Servers
- OS

You manage:
- Code only

### Examples:
- Azure App Service
- Google App Engine
- AWS Elastic Beanstalk

---

## 3. SaaS (Software as a Service)

👉 Fully ready-to-use software

You don’t manage anything

### Examples:
- Gmail
- Google Docs
- Microsoft 365

---

# 🧠 Easy Way to Remember

| Type | What You Manage |
|------|---------------|
| IaaS | Everything except hardware |
| PaaS | Only application code |
| SaaS | Nothing (just use it) |

---

# 🚀 Real-Life Analogy

👉 Renting a house:

- IaaS → Empty house (you manage everything inside)
- PaaS → Furnished house (just bring your stuff)
- SaaS → Hotel (everything ready)

---

# 🎯 Interview One-Liner

Cloud computing is the delivery of computing resources over the internet, and its main service models are IaaS, PaaS, and SaaS based on the level of control provided to the user.
