# 🐳 Docker Architecture

---

## 🧠 Overview

Docker follows a **client-server architecture**

```
Docker Client → Docker Daemon → Containers / Images → Registry
```

---

## 🖼️ Architecture Diagram (Concept)

```
+-------------------+        +----------------------+
|   Docker Client   | -----> |   Docker Daemon      |
| (docker commands) |        |   (dockerd)          |
+-------------------+        +----------+-----------+
                                        |
                     +------------------+------------------+
                     |                                     |
              +-------------+                      +----------------+
              |   Images    |                      |  Containers    |
              | (Templates) |                      | (Running Apps) |
              +-------------+                      +----------------+
                                        |
                                 +-------------+
                                 |  Registry   |
                                 | (DockerHub) |
                                 +-------------+
```

---

## 🧩 Components

### 1. 🖥️ Docker Client
- Where you run commands
- Example:
```bash
docker build
docker run
docker pull
```

---

### 2. ⚙️ Docker Daemon (dockerd)
- Background service
- Builds images
- Runs containers
- Manages Docker resources

---

### 3. 📦 Docker Images
- Read-only templates
- Created using Dockerfile

---

### 4. 🚀 Docker Containers
- Running instances of images
- Lightweight and isolated

---

### 5. 🌐 Docker Registry
- Stores Docker images
- Examples:
  - Docker Hub
  - AWS ECR
  - Azure ACR

---

## 🔄 Docker Workflow

```bash
# Build image
docker build -t myapp .

# Run container
docker run -p 5000:5000 myapp

# Push image
docker push myapp

# Pull image
docker pull myapp
```

# What is a Container (Technically)

## What is a Container technically?
A container is a lightweight, standalone, and executable package that includes:
- Application code
- Runtime
- System libraries
- Dependencies

### Key Concept: Layers of Images
Containers are built using **layered images**:
- Each layer represents a set of changes (e.g., installing packages, adding files)
- Layers are **read-only**
- A container adds a **writable layer** on top

#### Example Layers:
1. Base OS (Ubuntu / Alpine)
2. Installed packages (Python, Java)
3. Application code
4. Configuration

👉 This layered architecture makes containers:
- Fast
- Lightweight
- Reusable

---

## How Containers Work Internally
- Containers share the **host OS kernel**
- Isolation is achieved using:
  - Namespaces
  - Control Groups (cgroups)

```bash
docker pull nginx

# Docker vs Virtual Machine

## Introduction
Docker and Virtual Machines (VMs) are used to run applications in isolated environments. They differ in architecture, performance, and resource usage.

---

## Docker (Container-Based Virtualization)

### Architecture Diagram


--------------------------------------

| Applications (App + Dependencies + Libraries) |

--------------------------------------
| Docker Engine (Runtime) |

--------------------------------------
| Host OS Kernel |

--------------------------------------
| Infrastructure (VM / Physical) |

--------------------------------------



### Key Points
- Lightweight and fast
- Shares host OS kernel
- Starts in seconds
- Uses layered images
- Efficient resource utilization

---

## Virtual Machine (Hardware Virtualization)

### Architecture Diagram


--------------------------------------
| Applications |

--------------------------------------
| Guest OS (Full OS) |

--------------------------------------
| Hypervisor (VMware / KVM etc.) |

--------------------------------------
| Host OS |

--------------------------------------
| Infrastructure (Physical Server) |

--------------------------------------


### Key Points
- Heavyweight
- Each VM has its own OS
- Strong isolation
- Slower startup (minutes)
- Higher resource usage

---

## Docker vs VM Comparison

| Feature            | Docker (Containers) | Virtual Machine |
|--------------------|---------------------|-----------------|
| Virtualization     | OS-level            | Hardware-level  |
| OS                 | Shared              | Separate OS     |
| Size               | MBs                 | GBs             |
| Startup Time       | Seconds             | Minutes         |
| Performance        | High                | Moderate        |
| Isolation          | Process-level       | Full isolation  |
| Resource Usage     | Low                 | High            |

---

## Summary

Docker:
- Lightweight
- Fast
- Best for microservices and CI/CD

Virtual Machines:
- Strong isolation
- Full OS support
- Best for legacy and multi-OS workloads

---

## One-Line Difference

Docker = Lightweight containers sharing OS
VM = Full OS per instance with strong isolation



## ⚙️ How Docker Works Internally

### 1. Containers Share Host OS Kernel
- No separate OS per container
- Uses the host Linux kernel

### 2. Isolation Mechanism
Docker uses:
- **Namespaces** → isolate processes (PID, network, filesystem)
- **cgroups (Control Groups)** → limit CPU, memory, I/O

### 3. Layered Image System
Docker images are built in layers:
- Base Image (Ubuntu / Alpine)
- Installed packages (Python, Java)
- Application code

👉 Each layer is:
- Read-only
- Cached and reusable

Container = Image + Writable Layer

---

## 🚀 Key Advantages of Docker

- Very lightweight (MBs)
- Starts in seconds
- High performance (near-native)
- Easy to scale (microservices)
- Portable across environments

## ⚙️ How VM Works Internally

### 1. Hypervisor Layer
The hypervisor creates and manages virtual machines.

Types:
- Type 1 (Bare Metal): VMware ESXi, Hyper-V
- Type 2 (Hosted): VirtualBox, VMware Workstation

---

### 2. Each VM Has Its Own OS
- Full OS installed (Linux/Windows)
- Includes kernel + system processes

---

### 3. Resource Allocation
Each VM gets:
- Dedicated CPU
- RAM
- Storage

---

## 🚀 Key Advantages of VM

- Strong isolation (full OS separation)
- Can run different OS (Windows on Linux, etc.)
- Better security boundary
- Suitable for legacy systems

---

## ⚠️ Limitations of VM

- Heavy (GBs in size)
- Slow startup (minutes)
- High resource usage
- More overhead


# 🐳 Main Docker Commands

---

## 📥 docker pull

```bash
docker pull <image-name>
```

- Example:
```bash
docker pull nginx
```

---

## 🚀 docker run

```bash
docker run <image-name>
```

- Creates and starts a container

Example:
```bash
docker run nginx
```

## 📋 docker ps

```bash
docker ps
```

- Shows running containers

---

### Show all containers

```bash
docker ps -a
```

---

## ⛔ docker stop

```bash
docker stop <container-id>
```

- Stops a running container

---

## ▶️ docker start

```bash
docker start <container-id>
```

- Starts a stopped container

---

## 🌐 Port Mapping

```bash
docker run -p <host-port>:<container-port> <image>
```

- Maps container port to local machine

Example:
```bash
docker run -p 8080:80 nginx
```

👉 Access app at:
```
http://localhost:8080
```

---

## 🚀 Quick Workflow

```bash
docker pull nginx
docker run -p 8080:80 nginx
docker ps
docker stop <id>
docker start <id>
```

Pull → Run → View → Stop/Start → Map Ports

# 🐳 Additional Docker Commands (with Explanation)

---

## 📜 docker logs

```bash
docker logs <container-id>
```

### ✅ What it does:
- Displays **output (logs)** of a running or stopped container
- Useful for debugging errors

### 📌 Example:
```bash
docker logs my_container
```

---

## 🖥️ docker exec

```bash
docker exec -it <container-id> /bin/bash
```

### ✅ What it does:
- Lets you **enter inside a running container**
- Run commands as if you are inside the container

### 📌 Example:
```bash
docker exec -it my_container /bin/bash
```

---

## 🔍 docker inspect

```bash
docker inspect <container-id>
```

### ✅ What it does:
- Shows **detailed information** in JSON format
- Includes:
  - IP address
  - Port mappings
  - Environment variables
  - Configuration

### 📌 Example:
```bash
docker inspect my_container
```

---

## 🌐 docker network

### 🔹 List networks

```bash
docker network ls
```

### ✅ What it does:
- Shows all available Docker networks

---

### 🔹 Inspect network

```bash
docker network inspect <network-name>
```

### ✅ What it does:
- Shows details of a network
- Displays connected containers

---

### 🔹 Create network

```bash
docker network create my-network
```

### ✅ What it does:
- Creates a **custom network**
- Allows containers to communicate

---

## 🚀 Example Workflow

```bash
docker ps                  # see running containers
docker logs <id>           # check logs
docker exec -it <id> bash  # enter container
docker inspect <id>        # view details
docker network ls          # list networks
```

Logs → see output
Exec → go inside container
Inspect → full details

# 🐳 Dockerfile – Building Our Own Docker Image

---

## ❓ What is a Dockerfile?

- A **Dockerfile** is a script with instructions to build a Docker image
- It defines:
  - Base image
  - Dependencies
  - Commands to run the app

👉 Think:
```
Dockerfile = Recipe
Image = Final Dish
```

---

## 🧩 Basic Dockerfile Structure

```dockerfile
# Base image
FROM python:3.10

# Set working directory
WORKDIR /app

# Copy files
COPY . .

# Install dependencies
RUN pip install -r requirements.txt

# Expose port
EXPOSE 5000

# Run app
CMD ["python", "app.py"]
```

---

## 🛠️ What Each Instruction Does

### 🔹 FROM
- Defines base image
- Example:
```dockerfile
FROM python:3.10
```

---

### 🔹 WORKDIR
- Sets working directory inside container

---

### 🔹 COPY
- Copies files from local → container

---

### 🔹 RUN
- Executes commands (install dependencies)

---

### 🔹 EXPOSE
- Declares port used by container

---

### 🔹 CMD
- Default command to run app

---

## ✍️ Create Dockerfile

```bash
touch Dockerfile
```

👉 Add instructions as shown above

---

## 🏗️ Build Docker Image

```bash
docker build -t myapp .
```

### ✅ What it does:
- Reads Dockerfile
- Creates image named `myapp`

---

## 🚀 Run Container

```bash
docker run -p 5000:5000 myapp
```

### ✅ What it does:
- Starts container
- Maps port to localhost

---

## 🔄 Full Workflow

```bash
# Create Dockerfile
touch Dockerfile

# Build image
docker build -t myapp .

# Run container
docker run -p 5000:5000 myapp
```

---

## 🧠 Real Example

```
Flask App → Dockerfile → Image → Container → Running App
```

---

## 🎯 One-Line Summary

Dockerfile = instructions → build image → run container
