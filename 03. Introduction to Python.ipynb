⚙️ Interpreter vs Compiler

- Feature	Interpreter	Compiler
- Execution	Line by line	Whole program
- Speed	Slower	Faster
- Errors	Stops early	Shows all
- Output	No executable	Executable
- Example	Python, JS	C, C++

Python uses interpreter + bytecode (hybrid)
